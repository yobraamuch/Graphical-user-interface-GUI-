```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# (Q1) Loading our file
df=pd.read_excel("amazon.xlsx")
```

```python
# (Q1) First five records
df.head(5)
```

```python
# (Q2) Number of columns
df.columns
```

```python
# (Q2) Displaying the Datatypes
df.info()
```

```python
# (Q2) Number of rows and columns
df.shape
```

```python
# (Q2) Statistical summary of numerical columns
df.describe()
```

```python
# (Q3) Getting the total Sales
df['TotalAmount'].sum()
```

```python
df[['Quantity','UnitPrice','Discount','Tax','ShippingCost','TotalAmount']].head(10)
```

```python
df['Revenue']=df['Quantity']*df['UnitPrice']*(1-df['Discount'])
```

```python
df['Profit']=df['Revenue']-df['Tax']-df['ShippingCost']
```

```python
# (Q3) Calculating total profit
round(df['Profit'].sum(),2)
```

```python
# (Q3) Calculating Total Orders
df['OrderID'].count()
```

```python
# (Q3) Calculating Average Sales
round(df['TotalAmount'].mean(),2)
```

```python
# (Q3) Calculating Average Profit
round(df['Profit'].mean(),2)
```

```python
# (Q3) Calculating Maximum sales
round(df['TotalAmount'].max(),2)
```

```python
# (Q3) Calculating Minimum sales
round(df['TotalAmount'].min(),2)
```

```python
# (Q4) Total generated sales by each state
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=False).reset_index().round(2)
```

```python
# (Q5) Total sales by each category
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=True).reset_index()
```

```python
# (Q7) Total sales generated by each Sub-Category using ProductName
df.groupby(['Category','ProductName'])['TotalAmount'].sum().sort_values(ascending=True).reset_index()
```

```python
# (Q7) Total sales generated by each sub-category using Brand
df.groupby(['Category','Brand'])['TotalAmount'].sum().sort_values(ascending=True).reset_index()
```

```python
# (Q9) Top five Customers based on total sales
df.groupby(['CustomerName','ProductName'])['TotalAmount'].sum().sort_values(ascending=False).head(5).reset_index()
```

```python
# (Q10) Total Quantity sold for each product
df.groupby(['ProductName','Quantity','CustomerName'])['TotalAmount'].sum().sort_values(ascending=True).reset_index()
```

```python
# (Q11) Number of Orders placed using each paymentmethod
df.groupby('PaymentMethod')['OrderID'].count().sort_values(ascending=True).reset_index()
```

```python
# (Q12) Monthly sales trend using the Order Date
df['YearMonth']=df['OrderDate'].dt.to_period('M')
df.groupby('YearMonth')['TotalAmount'].sum().reset_index()
```

```python
# (Q13) The sales generated by each State for every Category
df.groupby(['State','Category'])['TotalAmount'].sum().reset_index()
```

```python
# (Q15) A bar chart showing sales by State
df.groupby('State')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
# (Q16) A pie chart showing Category-wise Sales
df.groupby('Category')['TotalAmount'].sum().plot(kind='pie')
```

```python
# (Q17) A line chart Showing Monthly Sales Trend
df.groupby('YearMonth')['TotalAmount'].sum().plot(kind='line')
```

```python
# (Q18) A horizontal Bar Chart showing the Top 10 Product by Sales
df.groupby('ProductName')['TotalAmount'].sum().head().plot(kind='barh')
```
